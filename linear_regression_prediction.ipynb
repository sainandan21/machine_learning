```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/Users/sainandanmamidi/Downloads/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Year)')
plt.ylabel('per capita')
per_capita_income_column = df['per capita income (US$)']
plt.scatter(df.year,per_capita_income_column, color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg = linear_model.LinearRegression()
reg.fit(df[['year']],per_capita_income_column)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year)')
plt.ylabel('per capita')
per_capita_income_column = df['per capita income (US$)']
plt.scatter(df.year,per_capita_income_column, color='red',marker='+')
plt.plot(df.year, reg.predict(df[['year']]), color="green")
```
